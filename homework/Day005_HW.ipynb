```python
import requests
import json

r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
response = r.text
data = json.loads(response) #type:list
print(type(data))
print(type(data[0]))
```

```python
# 1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
print('回傳幾筆資料:',len(data))
print('欄位:')
for key in data[0].keys():
    print(key)

```

```python
# 2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
import pandas as pd
title = ['title','createdAt','commentCount','likeCount']
context=[]
for d in data:
    context.append([d['title'],d['createdAt'],d['commentCount'],d['likeCount']])
df=pd.DataFrame(context,columns=title)
df
```
