## 作業目標

根據範例：

1. 取出今天所有的新聞
2. 取出現在時間兩小時內的新聞
3. 根據範例，取出三天前下午三點到五點的新聞

## 1. 取出今天所有的新聞

```python
# 打開瀏覽器
import re
import time
from selenium import webdriver
from bs4 import BeautifulSoup

browser = webdriver.Chrome(executable_path='./chromedriver')
browser.get("https://www.ettoday.net/news/news-list-2020-12-28-0.htm")

#Build Regex for today(2020/12/28)
regex = '2020/12/28\s\d+:\d+'
pattern = re.compile(regex)

# 每個兩秒鐘自動往下滑
time_pause=2
last_height = browser.execute_script("return document.body.scrollHeight")
while True:
    browser.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(time_pause)
    
    # 檢查是否成功取到頁面
    # Calculate new scroll height and compare with last scroll height
    new_height = browser.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        time.sleep(time_pause)
    last_height = new_height
    
    # BeautifulSoup 解析讀最後一筆時間 
    html_source = browser.page_source
    soup = BeautifulSoup(html_source, "html.parser")
    date = soup.find(class_="part_list_2").find_all('h3')[-1]
    test_date = date.find(class_="date").text
    if re.match(pattern, test_date):#時間為12/28繼續
        continue
    else:#時間非12/28 break
        break


#丟到 BeautifulSoup 解析   
for d in soup.find(class_="part_list_2").find_all('h3'):
    test_date = d.find(class_="date").text
    if re.match(pattern, test_date):# get news 12/28
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

## 2. 取出現在時間兩小時內的新聞

```python
#Build Regex for today(2020/12/28 within 2 hours)
regex2 = '2020/12/28\s1[8-9]:\d+'
pattern2 = re.compile(regex2)

for d in soup.find(class_="part_list_2").find_all('h3'):
    test_date = d.find(class_="date").text
    if re.match(pattern2, test_date):
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

```python
browser.quit();
```
