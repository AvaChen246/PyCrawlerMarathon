### 1. 根據範例使用 selenuim 取出 Hahow 網站中所有課程的資料，並用一個適合的資料結構儲存（Hint: 將所有資料存在一個變數當中取出）

```python
from selenium import webdriver
from selenium.webdriver.support.ui import Select
from bs4 import BeautifulSoup

browser = webdriver.Chrome(executable_path='./chromedriver')
browser.get("https://hahow.in/courses")
```

```python
# 取得資料（等瀏覽器資料出現後才執行）
html_source = browser.page_source
print(html_source)

# 取得資料
html_source = browser.page_source
soup = BeautifulSoup(html_source, 'html.parser')
soup.find('h4', class_='title').text
```

```python
course = []
courses = soup.find_all('h4', class_='title')
for c in courses:
    course.append(c.text)
print(course)
```

### 2. 剛剛的做法只能取出第一頁，試著取出所有頁面的資料。

```python
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select
from bs4 import BeautifulSoup


browser = webdriver.Chrome(executable_path='./chromedriver')
browser.get("https://hahow.in/courses")
```

```python
# 取得資料（等瀏覽器資料出現後才執行）
html_source = browser.page_source
soup = BeautifulSoup(html_source, 'html.parser')
```

```python
# 1. 先取出全部有幾頁
pages = soup.find('div', class_='pagination-container')
page = pages.find_all('li', class_='rc-pagination-item')
max_page=page[-1].text
max_page= int(max_page)
print(max_page)
```

```python
# 2. 每頁都取出資料放入 courses
import time
time_pause=10
course = []
for i in range(max_page):
    page=i+1
    browser.get("https://hahow.in/courses?page="+ str(page))
    #To check when the second page has finished loaded, another wait goes here
    #WebDriverWait(browser, time_pause).until(EC.presence_of_element_located((By.CLASS_NAME, 'pagination-container')))
    time.sleep(10)
    html_source = browser.page_source
    soup = BeautifulSoup(html_source, 'html.parser')
    courses = soup.find_all('h4', class_='course-title')
    for c in courses:
        course.append(c.text)
        
browser.close()
```

```python
# 3. done!

print(course)
```
